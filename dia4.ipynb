```python
# Diccionarios

contacto = {"nombre":"Analia",
            "apellido":"Britez",
            "correo":"analia@penguin.com"}

print(contacto)

print(contacto["nombre"]) # Llamar solo nombre

print(contacto.keys())  # claves
print(contacto.values()) # valores
print(contacto.items())  # claves + valores

# Añadir un elemento

contacto["teléfono"] = "0971123456"
print(contacto)

# Actualizar un elemento

contacto["teléfono"] = "0971098765"
print(contacto)

# Eliminar un elemento

contacto.popitem()
print(contacto)

# Eliminar un elemento específico

del contacto["apellido"]
print(contacto)
```

```python
# Challenge 2,1

tarjeta = True
tarjeta_personal = {}
while tarjeta: 
      

        print(f'''
Tarjeta Personal
        Elija un número:
        1. Crear
        2. Modificar
        3. Eliminar
        4. Salir
      ''')

        elección = input("Número: ")

        if elección == "1":
               
               newC = input("Nueva CLAVE: ")

               if newC in tarjeta_personal:
                     print("Este valor ya existe. Pruebe 'modificar'.")
                     print(f"Tu lista: {tarjeta_personal}")
               else:
                  newV = input("Nuevo VALOR:")

                  tarjeta_personal[newC] = newV
                  print(tarjeta_personal)

        elif elección == "2":
              
              modC = input("CLAVE EXISTENTE a cambiar: ")

              if modC not in tarjeta_personal:
                    print("Esta clave no existe. Pruebe 'crear'.")
                    print(f"Tu lista: {tarjeta_personal}")
              else:
                    modV = input("VALOR NUEVO a cambiar:")
                    tarjeta_personal[modC] = modV
                    print(tarjeta_personal)

        elif elección == "3":

            eliminar = input("CLAVE a eliminar")

            if eliminar not in tarjeta_personal:
                  print("Esta clave no existe.")
                  print(f"Tu lista: {tarjeta_personal}")
            else:
                  del tarjeta_personal[eliminar]
                  print(tarjeta_personal)
        
        elif elección == "4":
               print(f'''Esta es tu lista final:
                     {tarjeta_personal}''')
               print("Saliendo")
               break
        else:
              print("Número inválido.")
```

```python
# Challenge 3

agenda_prendida = True
agenda = {}
while agenda_prendida:
    
    print('''
        AGENDA DE CONTACTOS
      
      1. Mostrar los contactos
      2. Crear contacto
      3. Modificar datos del contacto
      4. Salir
      ''')
    
    
    elección = input("Número: ")

    if elección == "1":
      print(agenda)

    elif elección == "2":

      newC = input("Nuevo CONTACTO: ")

      if newC in agenda:

        print("Este contacto ya existe. Pruebe 'Modificar los datos del contacto'.")
        print(f"Tu lista: {agenda}")

      else:
        newT = input("Nuevo TELÉFONO:")
        newGmail = input("Nuevo CORREO:")
        agenda[newC] = newT, newGmail
        print(agenda)

    elif elección == "3":

      modC = input("CONTACTO EXISTENTE a cambiar: ")

      if modC not in agenda:
        print("Este contacto no existe. Pruebe 'Crear contacto'.")
        print(f"Tu lista: {agenda}")

      else:
        modT = input("TELÉFONO NUEVO a cambiar:")
        modGmail = input("CORREO NUEVO a cambiar:")
        agenda[modC] = modT, modGmail
        print(agenda)

    elif elección == "4":

      print(f'''Esta es tu agenda final:
            {agenda}''')
      print("Saliendo")
      break

    else:
      print("Número inválido.")
```
