```python
canasta_dulces = ["caramelo", "chicle", "paleta", "chocolate", "trufa"]

print(f'''              Esta es la canasta de dulces: 
    {canasta_dulces}
                    🍫  🍬  🍭  🍡
''')

# Agregar un elemento
print("Agregando un dulce nuevo...")
canasta_dulces.append("bombón")
print(canasta_dulces)
print(" ")

# Modificar el Segundo elemento (índice 1)
print("Modificando un dulce existente...")
canasta_dulces[1] = "turrón"
print(canasta_dulces)
print(" ")

# Eliminar el Tercer elemento (índice 2)
print("Retirando un dulce existente para comer xd")
del canasta_dulces[2]
print(canasta_dulces)
print(" ")
```
