```python
class Superheroe:
    def __init__(self, nombre):
        self.nombre = nombre
    
    def rayos_laser(self):
        print("Tiro láser de los ojos")

    def volar(self, km):
        print(f"Vuelo a {km} km por hora")

    def alzar(self, objeto, kg):
        print(f"Alzo 1 {objeto} de {kg} kg")

mi_héroe = Superheroe("Superman")
mi_héroe.rayos_laser()
mi_héroe.volar("25")
mi_héroe.alzar("casa", "500")
```

```python
class Auto:
    tipo = "Vehículo"

    def __init__(self, modelo, posición=0):
        self.modelo = modelo
        self.posición = posición

# FUNCIONES 

    def acelerar(self, metros):
        self.posición = metros + self.posición
        print(f"{self.modelo} acelera a {metros}m. Nueva posición: {self.posición}m")

    def girar(self, dirección):
        print(f"{self.modelo} gira a la {dirección}")

    def frenar(self):
        print(f"{self.modelo} frena y se detiene en la posición {self.posición}m")


miauto = Auto("RANGER RAPTOR")
miauto.acelerar(50)
miauto.girar("Derecha")
miauto.frenar()
```

```python
class Animal:
    tipo = "Animal Terrestre"

    def __init__(self, especie):
        self.especie = especie

    def ruido(self, sonido):
        print(f"Soy un {self.especie}. El ruido que hago es {sonido}!")

    def movimiento(self, acción):
        print(f"Al ser {self.especie}, cuando me muevo voy a {acción}.")

lobo = Animal("Lobo")
lobo.ruido("*aullido :v*")
lobo.movimiento("mi guarida")
print(" ")

leon = Animal("León")
leon.ruido("RAWRR")
leon.movimiento("cazar")
print(" ")
```

```python
class Dinosaurio:
    tipo = "Dinosaurio"

    def __init__(self, color, edad):
        self.color = color
        self.edad = edad

    def presentación(self):
        print(f"Soy un {self.tipo}")

    def rugido(self):
        print(f"RAWRR >:o")

class Pteorodactyl(Dinosaurio): 
    velocidad_vuelo = 70
    def volar(self):
        print(f"*vuela a {self.velocidad_vuelo}*")

class Velociraptor(Dinosaurio):
    velocidad_corrida = 40
    def correr(self):
        print(f"*corre a {self.velocidad_corrida}")

blue = Velociraptor("azul", "adulto")
blue.presentación()
blue.correr()

ave = Pteorodactyl("rojo", "infante")
ave.rugido()
ave.volar()
```

```python
class Dinosaurio:
    tipo = "Dinosaurio"

    def __init__(self, color, nombre, habilidad):
        self.color = color
        self.nombre = nombre
        self.habilidad = habilidad

    def presentación(self):
        print(f"Buenas!! Me llamo {self.nombre}.")

    def talento(self):
        print(f"Mi talento es {self.habilidad}!")

    def rugido(self):
        print(f"RAWRR >:o")

class Pterodactyl(Dinosaurio): 
    velocidad_vuelo = 70
    def volar(self):
        print(f"*vuela a {self.velocidad_vuelo}*")

class Velociraptor(Dinosaurio):
    velocidad_corrida = 40
    def correr(self):
        print(f"*corre a {self.velocidad_corrida}*")

blue = Velociraptor("azul", "Blue", "morder")
blue.presentación()
blue.talento()
blue.correr()

ave = Pterodactyl("rojo", "Ave", "cazar")
ave.presentación()
ave.rugido()
ave.talento()
ave.volar()
```
