```python
user_palabra = input("Ingrese una palabra")
user_num = int(input("Ingrese un número"))
def verificación(palabra, número):
    iguales = len(palabra) == número
    return iguales

Igualdad = verificación(user_palabra, user_num)
print(f"La igualdad '{user_palabra} = {user_num}' es: {Igualdad}.")
```
