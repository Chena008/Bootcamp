```python
total_caramelos = int(input("¿Cuál es tu cantidad de caramelos?:"))
niños = int(input("¿Cuántos niños hay?: "))
caramelos_por_niño = total_caramelos // niños

print(f"Hay {total_caramelos} caramelos y {niños} niños.")
print(f"Se entregaron {caramelos_por_niño} caramelos por niño")
caramelos_sobrantes = total_caramelos - (caramelos_por_niño * niños)
print(f"Sobraron {caramelos_sobrantes} caramelos")
```

```python
# Cafetería II >:D

#    user_product = input("¿Qué clase de café desea? ")
#    user_size = input("¿De qué tamaño desea su café? ")

def pedido():
    product = input("¿Qué clase de café desea? ")
    size = input("¿De qué tamaño desea su café? ")
    print(f'''Preparando su café de tipo "{product}"...
Cargando en vaso de tamaño {size}...''')
    print("     ¡Pedido listo!")
    print("     Disfrute :D ☕")
    print('''
          
          ''')


cantidad_pedidos = int(input('''¿Cuántos pedidos va a hacer?
    Proporcione un número del 1-4:  '''))

if cantidad_pedidos == 1:
    pedido()
elif cantidad_pedidos == 2:
        pedido()
        pedido()
elif cantidad_pedidos == 3:
     pedido()
     pedido()
     pedido()
elif cantidad_pedidos ==4: 
    pedido()
    pedido()
    pedido()
    pedido()
else:
     print("Lo siento, solo podemos hacer 1-4 pedidos.")
```
